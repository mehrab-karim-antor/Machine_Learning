LINEAR


```python
# Gradient Descent Example in Python (Colab Ready)

import numpy as np
import matplotlib.pyplot as plt

# Sample data (x vs y)
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2, 4, 6, 8, 10], dtype=float)  # y = 2x relationship

# Initialize parameters
m = 0.0   # slope
c = 0.0   # intercept
learning_rate = 0.01
epochs = 1000

# Gradient Descent loop
for i in range(epochs):
    Y_pred = m * X + c
    error = Y - Y_pred

    # Gradients
    dm = -(2/len(X)) * np.sum(X * error)
    dc = -(2/len(X)) * np.sum(error)

    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc

    # Print progress every 200 steps
    if i % 300 == 0:
        print(f"Epoch {i}: m={m:.4f}, c={c:.4f}")

print("\nFinal parameters:")
print(f"m = {m:.4f}, c = {c:.4f}")

# Plotting the result
plt.scatter(X, Y, color="blue", label="Data Points")
plt.plot(X, m*X + c, color="red", label="Fitted Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Gradient Descent Linear Regression")
plt.show()
```

NON_LINEAR


```python
# Non-linear Gradient Descent Example (Colab Ready)

import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**2 + 5*np.sin(x)

def df(x):
    return 2*x + 5*np.cos(x)

# Gradient Descent parameters
x = 3  # initial guess
learning_rate = 0.1
epochs = 70

history = []

for i in range(epochs):
    grad = df(x)
    x = x - learning_rate * grad
    history.append(x)
    print(f"Step {i}: x={x:.4f}, f(x)={f(x):.4f}")

print("\nFinal result:")
print(f"x = {x:.4f}, f(x) = {f(x):.4f}")

# Plot function and path
X = np.linspace(-6, 6, 400)
Y = f(X)

plt.plot(X, Y, label="f(x) = x^2 + 5sin(x)")
plt.scatter(history, [f(h) for h in history], color="red", label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on Non-linear Function")
plt.legend()
plt.show()
```
